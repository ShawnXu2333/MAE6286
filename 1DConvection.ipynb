## 1D linear convection

```python
import numpy as np
from matplotlib import pyplot as pp
%matplotlib inline
```

```python
# Set parameters
nx = 41
L = 2.0
dx = L / (nx-1)
nt = 25
dt = 0.02
c = 1.0
```

```python
x = np.linspace(0.0, L, num=nx)
x
```

```python
u0 = np.ones(nx)
mask = np.where((x >= 0.5)&(x <= 1.0))
mask
```

```python
u0[mask] = 2.0
```

```python
for i in range(nx):
    print(x[i],u0[i])
```

```python
def plot_figure(x, u0, u):
    pp.figure(figsize=(6.0,6.0))
    pp.xlabel('x')
    pp.ylabel('u')
    pp.grid()
    pp.plot(x, u0, color='b',label='Initial')
    pp.plot(x, u, color='r',label='Current')
    pp.legend()
    pp.xlim(0.0, 2.0)
    pp.ylim(0.0, 2.5)
```

```python
plot_figure(x, u0, u0)
```

```python
u = u0.copy()
for n in range(nt):
    un = u.copy()
    for i in range(1,nx):
        u[i] = un[i] - c * dt / dx * (un[i] - un[i-1])
```

```python
plot_figure(x, u0, u)
```

## 1D nonlinear convection

```python
u = u0.copy()
for n in range(nt):
    un = u.copy()
    for i in range(1,nx):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1])
```

```python
plot_figure(x, u0, u)
```

```python
u = u0.copy()
for n in range(nt):
    un = u.copy()
    for i in range(1,nx):
        u[i] = un[i] - un[i] * dt / dx * (un[i] - un[i-1])
    u[1:] = un[1:] - un[1:] * dt / dx * (un[1:] - un[:-1])
```

```python
plot_figure(x,u0,u)
```
